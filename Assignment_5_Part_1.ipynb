## It is better to rather use barplots when visualising categorical features. 
Therefore once again plot the model year and origin on two seperate graphs.
But instead using barplots this time.

```python
fig, ax = plt.subplots(figsize=(7,6))

# insert code here
categories = df['model_year'].value_counts().index
values = df['model_year'].value_counts()
ax.bar(categories, values, color ='#355C7D')
ax.set_xlabel('Model year', size=12)
ax.set_ylabel("Count", size =12)
ax.set_title('Model year counts', size =16)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(7,6))

# insert code here
category = df['origin'].value_counts().index
value = df['origin'].value_counts()
ax.bar(category, value, color ='#355C7D')
ax.set_xlabel("Origin", size=12)
ax.set_ylabel("Count", size =12)
ax.set_title("Count of cars based on origin", size=16)
plt.show()
```

## Comparing different years
Complete the code below to plot the temperature of each year on a different subplot.

```python
fig, ax = plt.subplots(10,1,figsize=(10,20))

# insert code here
years = pd.DatetimeIndex(df.index).year
eightyone = df['1981-01-01':'1981-12-31']
eightytwo = df['1982-01-01':'1982-12-31']
eightythree = df['1983-01-01':'1983-12-31']
eightyfour = df['1984-01-01':'1984-12-31']
eightyfive = df['1985-01-01':'1985-12-31']
eightysix = df['1986-01-01':'1986-12-31']
eightyseven = df['1987-01-01':'1987-12-31']
eightyeight = df['1988-01-01':'1988-12-31']
eightynine = df['1989-01-01':'1989-12-31']
ninety = df['1990-01-01':'1990-12-31']
    
def plot_timeseries(axes, x, y, color, xlabel, ylabel):
    axes.plot(x, y, color=color)
    axes.set_xlabel(xlabel)
    axes.set_ylabel(ylabel)
    
plot_timeseries(ax[0], eightyone.index, eightyone.iloc[:], '#C06C84', 'Year', 'Temperature (C)')
plot_timeseries(ax[1], eightytwo.index, eightytwo.iloc[:], '#C06C84', 'Year', 'Temperature (C)')
plot_timeseries(ax[2], eightythree.index, eightythree.iloc[:], '#C06C84', 'Year', 'Temperature (C)')
plot_timeseries(ax[3], eightyfour.index, eightyfour.iloc[:], '#C06C84', 'Year', 'Temperature (C)')
plot_timeseries(ax[4], eightyfive.index, eightyfive.iloc[:], '#C06C84', 'Year', 'Temperature (C)')
plot_timeseries(ax[5], eightysix.index, eightysix.iloc[:], '#C06C84', 'Year', 'Temperature (C)')
plot_timeseries(ax[6], eightyseven.index, eightyseven.iloc[:], '#C06C84', 'Year', 'Temperature (C)')
plot_timeseries(ax[7], eightyeight.index, eightyeight.iloc[:], '#C06C84', 'Year', 'Temperature (C)')
plot_timeseries(ax[8], eightynine.index, eightynine.iloc[:], '#C06C84', 'Year', 'Temperature (C)')
plot_timeseries(ax[9], ninety.index, ninety.iloc[:], '#C06C84', 'Year', 'Temperature (C)')

fig.suptitle('Daily Temperature Recorded for each respective year', size =16)
plt.tight_layout()
fig.subplots_adjust(top=0.95)
plt.show()
```

## Comparing distributions of time intervals
To properly visualise and compare the distributions of yearly or monthly data, we employ the use of boxplots.

### Write code below to plot a boxplot for the temperatures of each respective year.

```python
fig, ax = plt.subplots(figsize=(10,5))

# insert code here
eightyone = df['1981-01-01':'1981-12-31']
eightytwo = df['1982-01-01':'1982-12-31']
eightythree = df['1983-01-01':'1983-12-31']
eightyfour = df['1984-01-01':'1984-12-31']
eightyfive = df['1985-01-01':'1985-12-31']
eightysix = df['1986-01-01':'1986-12-31']
eightyseven = df['1987-01-01':'1987-12-31']
eightyeight = df['1988-01-01':'1988-12-31']
eightynine = df['1989-01-01':'1989-12-31']
ninety = df['1990-01-01':'1990-12-31']

ax.boxplot([eightyone.iloc[:],eightytwo.iloc[:],eightythree.iloc[:],
eightyfour.iloc[:],eightyfive.iloc[:],eightysix.iloc[:],
eightyseven.iloc[:],eightyeight.iloc[:],eightynine.iloc[:],
ninety.iloc[:]])
ax.set_xticklabels((pd.DatetimeIndex(df.index).year).unique())
ax.set_xlabel("Year", size=12)
ax.set_ylabel("Temperature (C)", size=12)
ax.set_title("Visualisations of the temperatures recorded from 1981 to 1990", size=16)
plt.tight_layout()
plt.show()
```
